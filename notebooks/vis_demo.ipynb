```python
import torch; torch.set_grad_enabled(False)
from transformers import AutoTokenizer, AutoModelForCausalLM

from delphi.eval.utils import load_orig_ds_txt, tokenize, get_correct_and_top_probs
from delphi.eval.vis import vis_sample_prediction_probs

model_name = "roneneldan/TinyStories-1M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
ds_txt = load_orig_ds_txt("validation[:100]")
ds_tok = [tokenize(tokenizer, txt) for txt in ds_txt]
sample_tok = ds_tok[0]

correct_probs, top_3_probs = get_correct_and_top_probs(model, sample_tok, top_k=3)
_, top_5_probs = get_correct_and_top_probs(model, sample_tok, top_k=5)
```

### collect top k predictions

```python
_ = vis_sample_prediction_probs(sample_tok, correct_probs, top_3_probs, tokenizer)
```

```python
_ = vis_sample_prediction_probs(sample_tok, correct_probs, top_5_probs, tokenizer)
```
