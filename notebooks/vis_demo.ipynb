```python
import torch; torch.set_grad_enabled(False)
from transformers import AutoTokenizer, AutoModelForCausalLM

from delphi.eval.utils import tokenize, get_next_and_top_k_probs, load_validation_dataset
from delphi.eval.vis import vis_sample_prediction_probs

model_name = "roneneldan/TinyStories-1M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
ds = load_validation_dataset("tinystories-v2-clean")
ds_txt = ds["story"][:100]
ds_tok = [tokenize(tokenizer, txt) for txt in ds_txt]
sample_tok = ds_tok[0]

correct_probs, top_3_probs = get_next_and_top_k_probs(model, sample_tok, k=3)
_, top_5_probs = get_next_and_top_k_probs(model, sample_tok, k=5)
```

### collect top k predictions

```python
_ = vis_sample_prediction_probs(sample_tok, correct_probs, top_3_probs, tokenizer)
```

```python
_ = vis_sample_prediction_probs(sample_tok, correct_probs, top_5_probs, tokenizer)
```
